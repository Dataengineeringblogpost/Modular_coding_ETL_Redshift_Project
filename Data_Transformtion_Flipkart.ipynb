```python
import pandas as pd
```

```python
Flipkart_Data = pd.read_csv("Your File directory")
Flipkart_Data.head(5)
```

```python
prefix = "Unnamed:"
cols_to_drop = Flipkart_Data.filter(like=prefix).columns
cols_to_drop
```

```python
Flipkart_Data.drop(columns=cols_to_drop,inplace = True)
Flipkart_Data
```

```python
# Extracting Brand Name out of Title
Flipkart_Data['Brand_Name'] = Flipkart_Data['Title'].str.split().str[0]
```

```python
Flipkart_Data['Brand_Name'].value_counts()
```

```python
#Droping the title column as its useless now
Flipkart_Data.drop("Title",inplace = True,axis=1)
```

```python
Flipkart_Data
```

```python
#Transform the processor column
Flipkart_Data['Processor'].value_counts()
```

```python
#Removing the genrations of the processor for easier classification
Flipkart_Data['Processor'] = Flipkart_Data['Processor'].str.split('(').str[0].str.strip()
Flipkart_Data['Processor'].value_counts()
```

```python
#Similarly u can replace the other rows as well
Flipkart_Data['Processor'] = Flipkart_Data['Processor'].str.replace('Intel Core 5 Processor', 'Intel Core i5 Processor')
Flipkart_Data['Processor'] = Flipkart_Data['Processor'].str.replace('Intel Core 7 Processor','Intel Core i7 Processor')
```

```python
Flipkart_Data['Processor'].value_counts()
```

```python
#Transform the RAM column
Flipkart_Data['RAM'].value_counts()
```

```python
# Function to extract the first 2 words (handles rows with less than 4 words)
def get_first_two_words(text):
  words = text.split(" ")
  return ' '.join(words[:2]) if len(words) >= 2 else text
```

```python
Flipkart_Data['RAM'] = Flipkart_Data['RAM'].apply(get_first_two_words)
```

```python
Flipkart_Data['RAM'].value_counts()
#Here we got the first two words of the RAM
```

```python
#Similarly u can replace the other rows as well similar to the one we did in the processor
Flipkart_Data['Operating_System'] = Flipkart_Data['Operating_System'].str.replace('64 bit Windows 11 Home Operating System', '64 bit Windows 11 Operating System')
```

```python
Flipkart_Data['Warranty'].value_counts()
```

```python
#Getting the first two words
Flipkart_Data['Warranty'] = Flipkart_Data['Warranty'].str.split().str[:2].apply(' '.join)
Flipkart_Data['Warranty'].value_counts()
```

```python
#Similarly u can replace the other rows as well
Flipkart_Data['Warranty'] = Flipkart_Data['Warranty'].str.replace('1 Yr', '1 Year')
Flipkart_Data['Warranty'] = Flipkart_Data['Warranty'].str.replace('1 Yera', '1 Year')
Flipkart_Data['Warranty'] = Flipkart_Data['Warranty'].str.replace('1 Years ', '1 Year')
Flipkart_Data['Warranty'] = Flipkart_Data['Warranty'].str.replace('1 year', '1 Year')
Flipkart_Data['Warranty'] = Flipkart_Data['Warranty'].str.replace('1 Years', '1 Year')
Flipkart_Data['Warranty'] = Flipkart_Data['Warranty'].str.replace('One-year International', '1 Year')
```

```python
Flipkart_Data['Warranty'].value_counts()
```

```python
Flipkart_Data['Warranty'] = Flipkart_Data['Warranty'].str.replace("3 Years",'3 Year')
Flipkart_Data['Warranty'] = Flipkart_Data['Warranty'].str.replace('2 Years','2 Year')
```

```python
Flipkart_Data['Warranty'].value_counts()
```

```python
#making the price column into a intiger column as it has comas and stuff

Flipkart_Data['Prices'] = Flipkart_Data['Prices'].str.replace(',', '')
```

```python
Flipkart_Data['Prices'] = Flipkart_Data['Prices'].str.replace(r'\D', '', regex=True)
```

```python
Flipkart_Data['Prices'] = pd.to_numeric(Flipkart_Data['Prices'], errors='coerce')

# Now you can try converting to integers if needed, but be aware that NaNs will remain
Flipkart_Data['Prices'] = Flipkart_Data['Prices'].astype('Int64', errors='ignore')
```
